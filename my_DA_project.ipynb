# Exploratory Data Analysis

# GENDER

```python
ax = sns.countplot(x = 'Gender',data = df , palette='Set1' )
ax.set_title("Gender vs No.of peoples")
# giving count on top
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# gender vs total amount
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' )
bx=sns.barplot(x = 'Gender' , y = 'Amount',data = sales_gen,palette = 'Set2')
bx.set_title("gender vs total amt")
for bars in bx.containers:
    bx.bar_label(bars)
```

conclusion :- by these analysis we conclude that most of purchases are done by women and purchase amount of women are more than men.

# Age

```python
ax=sns.countplot(x = 'Age Group' , data = df , palette = 'Set2', hue = 'Gender')
ax.set_title('age grp vs count')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_age = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' )
bx = sns.barplot(x = 'Age Group' , y = 'Amount', data = sales_age)
bx.set_title('age_grp vs amt')
```

from above analysis we can conclude that age_grp 26-35 shops more , and the shopping amount of this group is more than any other age group

# State

```python
df.columns
```

```python
# total no. of orders from the top 10 states
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False ).head(10)
ax=sns.barplot(x='State', y='Orders',data = sales_state)
ax.set_title('state vs orders')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
# total no. of amount from the top 10 states
state_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
bx=sns.barplot(x='State', y='Amount',data = state_amount)
bx.set_title('state vs Amount')
sns.set(rc={'figure.figsize':(15,5)})
```

Conclusion:- we can conclude that U.P, Maharastra and Karnataka were the top 3 ordering states with highest ordering amount 

# Marital Status

```python

ax = df.groupby(['Marital_Status'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' )
ax = sns.barplot(x = 'Marital_Status' , y = 'Orders' , data = ax)
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Marital_Status vs Orders')
```

```python
#Amount of married and unmarried buyers
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

Conclusion:- more no. of orders are placed by unmarried customers.

# Occupation

```python
x = df.groupby(['Occupation'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = True).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x = 'Occupation', y = 'Orders' ,data = x)
ax.set_title('Occupation vs Orders')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
x = df.groupby(['Occupation'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(x = 'Occupation', y = 'Orders' ,data = x)
ax.set_title('Occupation vs Orders')
for bars in ax.containers:
    ax.bar_label(bars)
```

Conclusion:- 
    Least no. of orders were placed by agriculture sector
    Most no. of orders were placed by IT sector and then by healthcare sector .

# Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category')
for bars in ax.containers:
    
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
ax.set_title('produt vs amount')
```

Conclusion :- most sold product category is clothing but the most amount generating product category is food. 

# Zone

```python
ax=df.groupby(['Zone'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' )
zones = ax['Zone'].tolist()
orders = ax['Orders'].tolist()
plt.pie(orders,labels = zones ,  autopct='%1.1f%%', startangle=90)
plt.title('orders from regions')
```

Conclsion :- Major buyers are from central and southern region of india.

# Conclusion

from our data analysis of diwali sales we conclude that unmarried womens between age_group 26-35yrs from central and southern zone ex. states like  UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category.
